# 使用方法

下面这个是必须要开头设定的：

````python
import stata_setup
stata_setup.config("/opt/Stata", "mp")
````

其中"/opt/Stata"是Stata的安装目录，可以自己设定；"mp"为版本，根据自己安装的版本设定

然后使用
```
%%stata
```
就可以运行stata代码了

````python
%%stata
sysuse auto, clear
sum
````

或者直接运行do文件：

````python
%%stata
do example.do
````
